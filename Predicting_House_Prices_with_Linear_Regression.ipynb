```python
def bi_varient_price(x):
  sns.barplot(data=df, x=x, y='price',color='brown',edgecolor='black')
  plt.xlabel(x.upper())
  plt.ylabel('PRICE IN MILLIONS')
  plt.title('RELATION OF PRICE AND '+str(x.upper()))
```
